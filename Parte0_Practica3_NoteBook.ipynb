**Dataset de viviendas**

Problema: Una empresa de aval√∫os y asesor√≠a inmobiliaria en Boston necesita una herramienta que le permita valuar viviendas.

```
import pandas as pd
from sklearn.datasets import load_boston
BostonDataset = load_boston()
df_Boston = pd.DataFrame(BostonDataset.data, columns=BostonDataset.feature_names)
df_Boston['target'] = pd.Series(BostonDataset.target)
df_Boston
```

**Tu propuesta**

El problema pide generar una herramienta para valuar viviendas. Por lo general, la valoraci√≥n de una vivienda se da como un precio, que es un valor espec√≠fico. Entonces este suena a un problema de regresi√≥n. Tenemos que ver los datos para confirmar que podemos hacer una regresi√≥n.

Al ver los datos, tenemos que preguntarnos:

1.   ¬øContamos con etiquetas o target? Si las tenemos podemos usar aprendizaje supervisado. Si no tenemos etiquetas, es aprendizaje no supervisado.
2.   ¬øDe qu√© tipo es la etiqueta o target? ¬øEs una clase discreta? ¬øEs un valor cont√≠nuo espec√≠fico? Si es una etiqueta de clase, entonces es clasificaci√≥n. Si es una etiqueta o target continuo (valor espec√≠fico), entonces es un problema de regresi√≥n.

Al ejecutar la celda de arriba, vemos el dataset y en la √∫ltima columna nos damos cuenta que el target es un valor cont√≠nuo. Por ejemplo, para la instancia 0, el valor del target es de 24.0; y para la instancia 4, de 36.2. As√≠ que confirmamos que es un problema de aprendizaje supervisado, y m√°s espec√≠ficamente, de regresi√≥n. Tambi√©n hemos ido a la documentaci√≥n del dataset para ver la descripci√≥n de las columnas.

S√≠ el dataset hubiera contenido un target con solo tres opciones: barato, medio, caro. Entonces este ser√≠a un problema de clasificaci√≥n, y nuestra herramienta para valuar s√≥lo tendr√≠a estas tres opciones.

```
X = df_Boston.iloc[:,0:-1] # Atributos: Todas las filas, de todas las columnas exceptuando la √∫ltima
y = df_Boston.iloc[:,-1]   # Etiqueta: √öltima columna

# Entrenar el modelo
from sklearn.svm import SVR   # Importar
regr = SVR(C=1.0, epsilon=0.2)# Crear
regr.fit(X, y)                # Entrenar o ajustar
```
