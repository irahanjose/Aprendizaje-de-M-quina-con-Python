## **Regresión**

Usemos un regresor basado en un árbol de decisión 🌳.

```
import numpy as np
import matplotlib.pyplot as plt

# Crear aleatoriamente datos sintéticos
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)    # X contiene un solo atributo: datos al azar
y = np.sin(X).ravel()                       # y contiene el valor real en función de x
y[::5] += 3 * (0.5 - rng.rand(16))          # Agregar ruido o valores incorrectos para no hacer tan fácil
                                            # el trabajo del regresor :P

# Crear y entrenar un modelo de regresión
from sklearn.tree import DecisionTreeRegressor  # Importar
regr = DecisionTreeRegressor()                  # Crear
regr.fit(X, y)                                  # Entrenar o ajustar

# Crear un arreglo de datos a lo largo del eje de las equis,
# y predecir el valor especifico del target (o "ye") con el regresor.
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_test = regr.predict(X_test)

# Graficar los resultados
plt.scatter(X, y, s=20, edgecolor="black", label="Datos")
plt.plot(X_test, y_test, label="Regresor | Árbol de Decisión", linewidth=2)
plt.xlabel("Datos")
plt.ylabel("target")
plt.legend()
```

Ahora usemos un regresor basado en una máquina de vectores de soporte.

```
#Importar módulos necesarios. Ya lo hemos hecho en la celda de arriba, pero para acostumbrarnos....
import numpy as np
import matplotlib.pyplot as plt

# Crear datos sintéticos aleatorios
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)      # X tiene un atributo
y = np.sin(X).ravel()                         # y contiene el valor real
y[::5] += 3 * (0.5 - rng.rand(16))            # Agregar ruido

# Entrenar el modelo
from sklearn.svm import SVR  # Importar
regr = SVR()                 # Crear
regr.fit(X, y)               # Entrenar o ajustar

# Predecir
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_test = regr.predict(X_test)

# Graficar los resultados
plt.scatter(X, y, s=20, edgecolor="black", label="Datos")
plt.plot(X_test, y_test, label="Regresor | Vectores de Soporte", linewidth=2)
plt.xlabel("Datos")
plt.ylabel("target")
plt.legend()
```
