LLegamos al paso relacionado con el **entrenamiento del modelo** üí™. Para este paso, usaremos [scikit-learn](https://es.wikipedia.org/wiki/Scikit-learn) (tambi√©n conocido como sklearn). Sklearn es de software libre, y contiene funciones para aprendizaje de m√°quina en Python.

Los pasos para usar un modelo y entrenarlo son:
1. Importar el modelo de sklearn
2. Crear el modelo
3. Entrenar el modelo

```
# Usar y entrenar un modelo: √Årbol de decisi√≥n
from sklearn.tree import DecisionTreeClassifier # Importar el modelo/algoritmo de sklearn
clf = DecisionTreeClassifier()                  # Crear el modelo
```

¬øCu√°l es el siguiente paso? ... Exacto ‚ú®, entrenar el modelo.

```
# Para entrenar el modelo, usaremos la funci√≥n fit. Esta funci√≥n toma dos par√°metros.
# El primero es un arreglo con los atributos, y el segundo es un arreglo con las etiquetas de las clases.

# Como nuestros datos est√°n en un dataframe que contiene tanto los atributos como las etiquetas, primero
# necesitamos separarlos en dos arreglos. Esto es f√°cil usando iloc en pandas.

# Los atributos son los valores de los ejemplos, y la etiqueta es el valor o clase que intentamos predecir
X = df.iloc[:,0:-1] # Atributos: Todas las filas, de todas las columnas exceptuando la √∫ltima
y = df.iloc[:,-1]   # Etiqueta: √öltima columna, 'CreditScore' 0: Negar pr√©stamo; 1: Asignar pr√©stamo

# Ahora s√≠ podemos entrenar el modelo con la funci√≥n fit(X,y)
clf = clf.fit(X, y)
```

Listo, ya tienes un modelo entrenado. Asignemos una calificaci√≥n a este modelo. En la pr√°ctica 4, hablaremos m√°s sobre la evaluaci√≥n de modelos de aprendizaje.

```
# Evaluar el modelo: √Årbol de decisi√≥n
from sklearn.metrics import accuracy_score  
y_pred = clf.predict(X)                             # Usar los ejemplos de prueba para predecir

print('Calificaci√≥n: ', accuracy_score(y, y_pred))  # Comparar las predicciones con los valores
                                                    # reales de las etiquetas de prueba
```

Ejecuta las siguientes celdas para entrenar y evaluar dos modelos m√°s.

```
# Entrenar un modelo: Gaussian N√§ive Bayes
from sklearn.naive_bayes import GaussianNB  # Importar el modelo de sklearn
clf = GaussianNB()                          # Crear el modelo
clf = clf.fit(X, y)                         # Entrenar el modelo. Recuerda: En una celda anterior ya hemos
                                            # asignado a "X" los atributos, y a "y" las etiquetas de las clases

# Evaluar el modelo: Gaussian N√§ive Bayes
y_pred = clf.predict(X)                # usar los ejemplos de prueba para predecir

print('Calificaci√≥n: ', accuracy_score(y, y_pred))  # Comparar las predicciones con los valores
                                                    # reales de las etiquetas de prueba
```

```
# Entrenar un modelo: M√°quina de Vectores de Soporte SVM
# Equivalente a  from sklearn.svm import SVC
from sklearn import svm               # Importar el modelo de sklearn
clf = svm.SVC()                       # Crear el modelo
clf = clf.fit(X, y)                   # Entrenar el modelo

# Evaluar el modelo: M√°quina de Vectores de Soporte SVM
y_pred = clf.predict(X)          # usar los ejemplos de prueba para predecir

print('Calificaci√≥n: ', accuracy_score(y, y_pred))  # Comparar las predicciones con los valores
                                                    # reales de las etiquetas de prueba
```
